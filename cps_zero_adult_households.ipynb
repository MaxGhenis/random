# How many zero-adult households are in the CPS?

## Setup

```
import pandas as pd
import numpy as np
```

### Load data

CSV result of this IPUMS CPS extract:

![img](https://imgur.com/CpSUdqK.png)

```
# https://drive.google.com/file/d/1OvkchWDqrLWSAaW96ZmuwasH4RgYM5z8
cps_raw = pd.read_csv('https://drive.google.com/uc?export=download&id=1OvkchWDqrLWSAaW96ZmuwasH4RgYM5z8',
                      compression='gzip')
```

## Preprocess

```
cps = cps_raw.copy(deep=True)
```

```
cps.columns = map(str.lower, cps.columns)
```

```
cps['nu18'] = cps.age < 18
cps['n18'] = cps.age >= 18
```

```
cps_hh = cps.groupby('cpsid')[['n18']].sum()
cps_hh.columns = ['n18_hh']
```

```
cps = cps.merge(cps_hh, left_on='cpsid', right_index=True)
```

## Analysis

```
cps.groupby('year')[['asecwt', 'asecwth']].sum() / 1e6
```

```
cps[cps.n18_hh == 0]
```

```
cps['asecwth_zero_adults'] = np.where(cps.n18_hh == 0, cps.asecwth, 0)
```

```
cps.groupby('year')[['asecwth', 'asecwth_zero_adults']].sum()
```
